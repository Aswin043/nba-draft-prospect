# EDA


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV
df = pd.read_csv("nba_draft_2025_final.csv")

# Convert stat columns to numeric (if not already)
stat_cols = ["fg_pct", "fg3_pct", "ft_pct", "mp_per_g", "pts_per_g", "trb_per_g", "ast_per_g"]
df[stat_cols] = df[stat_cols].apply(pd.to_numeric, errors='coerce')
```

```python
print(df[stat_cols].describe())
```

```python
df[stat_cols].hist(bins=15, figsize=(15, 10), color="#6495ED")
plt.suptitle("Distribution of College Stats")
plt.tight_layout()
plt.show()
```

```python
# Compute correlations between stats
corr_matrix = df[stat_cols].corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between College Stats")
plt.show()
```

```python
sns.pairplot(df[stat_cols])
plt.suptitle("Pairwise Relationships Between Stats", y=1.02)
plt.show()
```
